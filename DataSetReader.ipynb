```python
import pandas as pd
import concurrent.futures


chunksize = 10000
json_reader = pd.read_json('yelp_academic_dataset_review.json', lines=True, chunksize=chunksize)

df_list = []
for chunk in json_reader:
    df_list.append(chunk)
    
# Now df_list is a list of dataframes, which you can concatenate to get a single dataframe
reviewDF = pd.concat(df_list)

display(reviewDF)
```

```python
chunksize = 10000
json_reader = pd.read_json('yelp_academic_dataset_business.json', lines=True, chunksize=chunksize)

df_listBusiness = []
for chunk in json_reader:
    df_listBusiness.append(chunk)
    
# Now df_list is a list of dataframes, which you can concatenate to get a single dataframe
businessDF = pd.concat(df_listBusiness)

display(businessDF)
```

```python
merged_df = pd.merge(reviewDF, businessDF, on='business_id', how='outer')
```

```python
display(merged_df)
```

```python
filtered_df = merged_df[merged_df['categories'].str.contains('Restaurant', case=False, na=False)]
filtered_df = filtered_df[filtered_df['city'].str.contains('Philadelphia', case=False, na=False)]

display(filtered_df)
```

```python
import pandas as pd
import json

# Assuming merged_df is your dataframe
grouped = filtered_df.groupby('business_id')

for business_id, group in grouped:
    # Getting business details - assuming 'name' is a column with business name
    # Add other columns as needed
    business_details = group[['name', 'business_id', 'categories']].iloc[0].to_dict()
    
    # Getting reviews - assuming 'review_text' is the column with the review text
    # Adjust the column name based on your dataframe
    reviews = group['text'].tolist()
    
    # Creating a dictionary with business details and reviews
    data = {
        'business_details': business_details,
        'reviews': reviews
    }
    
    # Creating a unique filename using the business ID
    filename = f'DataCollection/DataAsJsons/business_{business_id}.json'
    
    # Saving the data to a JSON file
    with open(filename, 'w') as f:
        json.dump(data, f, indent=4)
```
